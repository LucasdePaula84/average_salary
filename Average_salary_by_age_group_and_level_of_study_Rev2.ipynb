# Carregamento de dados e pacotes

```python
import pandas as pd
import matplotlib.pyplot as plt


# Caminho do arquivo CSV
caminho_arquivo = "Documents/Python/Media_Salarial/salary_prediction_data.csv"

# Lendo o arquivo CSV
dados = pd.read_csv(caminho_arquivo)

# Apresentando as primeiras linhas do arquivo
dados.head()
```

```python
# Verificando os tipos de dados por coluna

print("Tipos de dados por coluna:")
dados.info()
```

# 1. Analises Estatistica:

```python
# Resumo estatístico das variáveis numéricas

print("Resumo estatístico das variáveis numéricas:")
print(dados[['Experience', 'Age', 'Salary']].describe())
```

```python
# Contagem de valores únicos para variáveis categóricas

print("\nContagem de valores únicos para variáveis categóricas:")
print("Educação:")
print(dados['Education'].value_counts())
print("\nLocalização:")
print(dados['Location'].value_counts())
print("\nTítulo do cargo:")
print(dados['Job_Title'].value_counts())
print("\nGênero:")
print(dados['Gender'].value_counts())
```

```python


# Distribuição de salários por nível de educação
plt.figure(figsize=(10, 6))
dados.groupby('Education')['Salary'].mean().plot(kind='bar')
plt.title('Distribuição de salários por nível de educação')
plt.xlabel('Nível de Educação')
plt.ylabel('Salário Médio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Distribuição de salários por localização
plt.figure(figsize=(10, 6))
dados.groupby('Location')['Salary'].mean().plot(kind='bar')
plt.title('Distribuição de salários por localização')
plt.xlabel('Localização')
plt.ylabel('Salário Médio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Distribuição de salários por título do cargo
plt.figure(figsize=(12, 6))
dados.groupby('Job_Title')['Salary'].mean().plot(kind='bar')
plt.title('Distribuição de salários por título do cargo')
plt.xlabel('Título do Cargo')
plt.ylabel('Salário Médio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Distribuição de salários por sexo
plt.figure(figsize=(12, 6))
dados.groupby('Gender')['Salary'].mean().plot(kind='bar')
plt.title('Distribuição de salários por sexo')
plt.xlabel('Sexo')
plt.ylabel('Salário Médio')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Distribuição de idades por nível de educação
plt.figure(figsize=(10, 6))
dados.boxplot(column='Age', by='Education')
plt.title('Distribuição de idades por nível de educação')
plt.xlabel('Nível de Educação')
plt.ylabel('Idade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Distribuição de idades por localização
plt.figure(figsize=(10, 6))
dados.boxplot(column='Age', by='Location')
plt.title('Distribuição de idades por localização')
plt.xlabel('Localização')
plt.ylabel('Idade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Distribuição de idades por título do cargo
plt.figure(figsize=(12, 6))
dados.boxplot(column='Age', by='Job_Title')
plt.title('Distribuição de idades por título do cargo')
plt.xlabel('Título do Cargo')
plt.ylabel('Idade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 2. Analise de Correlação

```python
# Criando gráficos de dispersão separados para cada variável
plt.figure(figsize=(18, 6))

# Gráfico de dispersão para 'Experience' vs 'Salary'
plt.subplot(1, 3, 1)
plt.scatter(dados['Experience'], dados['Salary'], color='blue')
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.grid(True)

# Gráfico de dispersão para 'Age' vs 'Salary'
plt.subplot(1, 3, 2)
plt.scatter(dados['Age'], dados['Salary'], color='red')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)

# Gráfico de dispersão para 'Salary' vs 'Salary' (apenas para criar espaço para a legenda)
plt.subplot(1, 3, 3)
plt.scatter(dados['Salary'], dados['Salary'], color='green', label='Salary')
plt.title('Salary vs Salary')
plt.xlabel('Salary')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
